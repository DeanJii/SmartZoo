```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
result_Y = np.load("./drive/MyDrive/Tutoring/duck_Y_sequence.npy")
result_X = np.load("./drive/MyDrive/Tutoring/duck_X_sequence.npy")
```

```python
rawdata = pd.read_csv("./drive/MyDrive/Tutoring/Lake_Constance_Ducks_XY.csv", encoding='CP949') # if you want to use another animal data, you can change this line

real = rawdata.dropna().to_numpy()
generated = np.c_[result_X, result_Y]
```

```python
## real + generated data ##
## hint: data = np.concatenate((real_sampled, generated), axis = 0)
data = np.concatenate((real, generated), axis = 0)
```

```python
data
```

```python
# scaling
min_max_scaler = MinMaxScaler()
data_scaled = min_max_scaler.fit_transform(data.reshape(-1,1))
data_scaled = data_scaled.reshape(-1, 2)
```

```python
# Plot
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], s=20, c='y', marker='o')
plt.title('Generated + real movement data of duck (scaled)')
plt.legend(loc=2)
plt.show()
```

```python
# scaling
min_max_scaler = MinMaxScaler()
real_scaled = min_max_scaler.fit_transform(real.reshape(-1,1))
real_scaled = real_scaled.reshape(-1, 2)
```
